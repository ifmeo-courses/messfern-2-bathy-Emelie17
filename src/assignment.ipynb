## Writing Your First Function

Now that we have loaded the bathymetry data, let's learn to write a **function**! Functions are reusable pieces of code that take inputs, do something useful, and return results.

We'll write a function to extract the water depth at any location. This will be useful for finding depths at specific points like the DS2 mooring location.

```python
# YOUR CODE HERE
# Write a function to extract water depth at a specific location
def get_depth_at_location(bathymetry_dataset, target_lat, target_lon):
    """
    Extract water depth at a specific location.
    
    Parameters:
    - bathymetry_dataset: xarray Dataset with lat, lon, z variables
    - target_lat: Target latitude (degrees North)  
    - target_lon: Target longitude (degrees East, negative for West)
    
    Returns:
    - depth: Water depth at location (meters, negative below sea level)
    """
    # HINT: Use bathymetry_dataset.z.sel(lat=target_lat, lon=target_lon, method='nearest')
    # This finds the nearest grid point to your target location.  Call it `depth_value`
    depth_value=bathymetry_dataset.z.sel(lat=target_lat, lon=target_lon, method='nearest')

    # Then extract the single depth value using .values or .item()
    # Extract the single value (handle both scalar and array returns)
    if hasattr(depth_value, 'item'):
        return depth_value.item()
    else:
        return float(depth_value.values)

# Test your function with the DS2 mooring location
ds2_latitude = 66.0128
ds2_longitude = -27.270200

# Call your function to get the depth at DS2 mooring
ds2_depth = get_depth_at_location(bathymetry_subset, ds2_latitude, ds2_longitude)
print(f"Water depth at DS2 mooring ({ds2_latitude}°N, {ds2_longitude}°E): {ds2_depth:.1f} m")
```

Further away from the equator the distortion gets bigger. It is better to use cartopy to represent the region geographically correctly, because it applies a proper geographic projection in which the proportions are shown correctly. In addition to that you can easily add coastlines and meridians and similar features. To plot the bathymetry and show the ocean seafloor features in a representetive way it is necessary that the proportions are shown correctly. 

2. Because the earth is round it is hard to show it in 2D. When you do a map projection you can never depict the correct distances, angles, areas and forms all at the same time. With bathymetry data ocean seafloor features like mid-ocean ridges, trenches or continental slopes are shown. The North Atlantic is in the mid latitudes so the distortion is maybe still acceptable. Second the Mercator projection shows the angles correctly which is important for coastlines and for example mid ocean ridges. And another point is that the Mercator projection is conformal but not equal-area. That means the structures are plotted as they actually look but maybe the size of what is shown is scaled.
